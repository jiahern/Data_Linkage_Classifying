```python
import csv
import pandas as pd
import numpy as np
import textdistance as td
from collections import defaultdict


df_amazon_sml = pd.read_csv("amazon_small.csv")
df_google_sml = pd.read_csv("google_small.csv")
df_amazon_ggl_sml = pd.read_csv("amazon_google_truth_small.csv")
df_amazon = pd.read_csv("amazon.csv")
df_amazon_ggl = pd.read_csv("amazon_google_truth.csv")
df_google = pd.read_csv("google.csv")
df_yeast = pd.read_csv("all_yeast.csv")
```

```python

df_google_sml['description'] = df_google_sml['description'].replace(np.nan, " ",regex=True)
df_google_sml['name_desc'] = df_google_sml['name'] + df_google_sml['description']
print(df_google_sml['name_desc'])

df_amazon_sml['description'] = df_amazon_sml['description'].replace(np.nan, " ",regex=True)
df_amazon_sml['name_desc'] = df_amazon_sml['title'] + df_amazon_sml['description']
print(df_amazon_sml['name_desc'])
```

```python
scores = []

df_google_sml['name_desc'] = df_google_sml['name'] + df_google_sml['description']
df_amazon_sml['name_desc'] = df_amazon_sml['title'] + df_amazon_sml['description']

for item in df_amazon_sml["name_desc"]:
    stripped_amazon = item
    stripped_amazon = ''.join([c for c in stripped_amazon if c not in (':',',','(',')','[',']','&','-','--','!',';')])
    tokens_1 = stripped_amazon.split()
    tokens_1
    for item1 in df_google_sml["name_desc"]:
        stripped_google = item1
        stripped_google = ''.join([c for c in stripped_google if c not in (':',',','(',')','[',']','&','-','--','!',';')])
        tokens_2 = stripped_google.split()
        sim_score = td.jaccard(tokens_1, tokens_2)
        scores.append(sim_score)

print(scores)


def split(arr, size):
     arrs = []
     while len(arr) > size:
         pice = arr[:size]
         arrs.append(pice)
         arr   = arr[size:]
     arrs.append(arr)
     return arrs
split_scores = split(scores,150)
print(split_scores[1])
list_count = 0
num_count = 0

max_values = []

# finds the maximum similarity for each comparison CHANGE WORDING 
for lst in split_scores:
    max_values.append(max(lst))

position = []
for num in max_values:
    for lst in split_scores:
        for sim in lst:
            if sim == num:
                position.append(lst.index(num))
print(len(position))

position = {}
for value in max_values:
    for lists in split_score:
        for i in range(len(df_google_sml)):
            if value == lists[i]:
                index = lists.index(lists[i])
                position[value] = index
            else:
                continue
print(len(position.values()))

final = []
for value in max_values[0:4]:
    for pos in position[0:4]:
        final.append(value)
        final.append(pos)
print(final)

        
```

```python
# comparing classification algorithms
from sklearn.tree import DecisionTreeClassifier
#reference: https://towardsdatascience.com/building-a-k-nearest-neighbors-k-nn-model-with-scikit-learn-51209555453a
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
import pandas as pd
import matplotlib.pyplot as plt

from sklearn import neighbors
from sklearn.metrics import accuracy_score
from sklearn import preprocessing


classifier_data = median_replaced
##k = 5
#storing classifiers and targets
X = classifier_data.drop(columns=["Class", "Sample"])
y = classifier_data["Class"].values
X.head()

#splitting dataset into training and test data 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=1/3, random_state=1, stratify=y)

# Create KNN classifier for k = 5
knn_5 = KNeighborsClassifier(n_neighbors = 5)

# Fit the classifier to the data
knn_5.fit(X_train,y_train)

#show first 8 model predictions on the test data using mcg-nuc
y_pred_5 = knn_5.predict(X_test)
print("predicted Class features of data when k = 5:", knn_5.predict(X_test))


##k = 10
#storing classifiers and targets
X = classifier_data.drop(columns=["Class"])
y = classifier_data["Class"].values

# Create KNN classifier for k = 10
knn_10 = KNeighborsClassifier(n_neighbors = 10)

# Fit the classifier to the data
knn_10.fit(X_train,y_train)

#show first 8 model predictions on the test data using mcg-nuc
print("Predicted Class features of data, when k = 10:", knn_10.predict(X_test))
y_pred_10 = knn_10.predict(X_test)


dtree = DecisionTreeClassifier(max_depth = 10, random_state = 101, max_features = None, min_samples_leaf = 15)
dtree.fit(X_train, y_train)
y_pred_tree = dtree.predict(X_test)
print("Decision tree accuracy:", accuracy_score(y_test, y_pred_tree))
print("k = 5 accuracy:", accuracy_score(y_test, y_pred_5 ))
print("k = 10 accuracy:", accuracy_score(y_test, y_pred_10 ))
```
